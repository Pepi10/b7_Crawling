### '사원수', '매출액', '직급' 열의 데이터 타입을 수치형(서수형)으로 변경


사원수

```python
# '사원수' 열의 데이터 타입을 문자열로 변환
pro_end_df['사원수'] = pro_end_df['사원수'].astype(str)

# '사원수' 열에서 숫자를 제외한 모든 문자를 제거하고 정수형으로 변환
pro_end_df['사원수'] = pro_end_df['사원수'].str.replace(r'\D', '', regex=True).astype(int)

# 변경된 데이터프레임 확인
print(pro_end_df['사원수'].dtype)
```

매출액

```python
import numpy as np

# 매출액 열에서 숫자를 제외한 모든 문자를 제거하고 수치형(연속형)으로 변환
pro_end_df['매출액'] = pro_end_df['매출액'].str.replace(r'[^0-9]', '', regex=True)

# 빈 문자열을 NaN 값으로 대체
pro_end_df['매출액'] = pro_end_df['매출액'].replace('', np.nan)

# NaN 값을 가진 행 삭제
pro_end_df = pro_end_df.dropna(subset=['매출액'])

# 데이터 타입을 float로 변환
pro_end_df['매출액'] = pro_end_df['매출액'].astype(float)

# 만원 단위로 변환
pro_end_df['매출액'] = pro_end_df['매출액'] / 10000

# 변경된 데이터프레임 확인
print(pro_end_df['매출액'])

```

직급(잠시 접어둬요.. 경력무관을 0으로 바꾸는 거에서 막힘)

```python
import numpy as np

# '경력무관'을 0으로 대체
pro_end_df['직급'] = pro_end_df['직급'].replace('경력무관', 0)

# '직급' 열에서 '년'을 제거하고 최소 조건을 기준으로 수치형(서수형)으로 변경
def extract_min_years(x):
    if isinstance(x, str):
        if ' ~ ' in x:
            return float(x.split(' ~ ')[0].replace('년', '').strip())
        elif '이상' in x:
            return float(x.replace('년 이상', '').strip())
    return np.nan

pro_end_df['직급'] = pro_end_df['직급'].apply(extract_min_years)

# 변경된 데이터프레임 확인
print(pro_end_df['직급'])
```
