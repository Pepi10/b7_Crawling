```python
import pandas as pd
import re

# 엑셀 파일 경로
file_path = '../EDA자료/pro_end.xlsx'

# 엑셀 파일 읽기
df = pd.read_excel(file_path)

# 주소에서 구, 도, 시를 추출하는 함수
def extract_province_city_district(address):
    # 주소에 "서울"이 없는 경우 "서울특별시"로 대체
    if "서울" not in address:
        address = "서울특별시 " + address

    # 주소에 "인천"이 없는 경우 "인천광역시"로 대체
    if "인천" not in address:
        address = "인천광역시 " + address
    
    # 주소에서 "구"를 추출
    district_match = re.search(r"(\S+[구군])", address)
    district = district_match.group() if district_match else None
    
    # 주소에서 "도"를 추출
    province_match = re.search(r"(\S+[도])", address)
    province = province_match.group() if province_match else None

    # 주소에서 "시"를 추출
    city_match = re.search(r"(\S+[시도])", address)
    city = city_match.group() if city_match else None

    

    return province, city, district

# 주소 열에서 도, 시, 구를 추출하여 새로운 열에 저장
df['도'], df['시'], df['구'] = zip(*df['주소'].apply(extract_province_city_district))

# 전처리된 데이터 확인
print(df[['주소', '도', '시', '구']])

# 전처리된 데이터 저장
processed_file_path = "processed_pro_end.xlsx"
df.to_excel(processed_file_path, index=False)

print("전처리된 데이터가 저장되었습니다:", processed_file_path)
```
