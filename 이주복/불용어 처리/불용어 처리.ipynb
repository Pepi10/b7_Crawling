```python
# 한글 명사 분석
tec_word = ''
try:
    for j in data['이용하는기술스택/우대사항']:       
        tec_word += ' '.join(kkma.nouns(j.replace('ㆍ','')))
except:
    pass

tec_word
```

```python
# 영어 명사 분석
abc = ''
try:
    for q in data['이용하는기술스택/우대사항']:
        abc += ''.join(re.sub('[^a-zA-Z]', ' ', q))
except:
    pass

abc
```

```python
# 한글, 영어 명사 합치기
abc_sum = abc + tec_word
abc_sum
```
